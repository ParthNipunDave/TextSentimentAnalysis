```python

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.metrics.pairwise import sigmoid_kernel
from flask import Flask
app=Flask(__name__)
PropertDataSet =pd.DataFrame( pd.read_json("P:/Data Science/webmob/Property.json"))

UserDataSet =pd.DataFrame( pd.read_json("P:/Data Science/webmob/User.json"))

UserDataSet = UserDataSet.loc[:,["_id","userSaveHomes"]]

PropertDataSet = PropertDataSet.loc[:,["_id","bedroom","bathroom","tags","price"]]

PropertDataSet["tags"] = PropertDataSet["tags"].astype(str)
PropertDataSet["_id"] = PropertDataSet["_id"].astype(str)
PropertDataSet["bedroom"] = PropertDataSet["bedroom"].astype(str)
PropertDataSet["bathroom"] = PropertDataSet["bathroom"].astype(str)


tf = TfidfVectorizer( max_features=None,
                             min_df=0,strip_accents="unicode",
                        analyzer="word",
                      token_pattern=r"\w{1,}",
                             ngram_range=(1,3))


print(PropertDataSet.shape)

#fitting and transforming matrix
bedMat = tf.fit_transform((PropertDataSet["bedroom"]))
bathMat = tf.fit_transform((PropertDataSet["bathroom"]))
tagsMat = tf.fit_transform((PropertDataSet["tags"]))
priceMat = tf.fit_transform((PropertDataSet["price"]))

#Finding sigmoids

bathSig = sigmoid_kernel(bathMat,bathMat)
bedSig = sigmoid_kernel(bedMat,bedMat)
tagsSig= sigmoid_kernel(tagsMat,tagsMat)
priceSig = sigmoid_kernel(priceMat,priceMat)

#indexs

indice = pd.Series(PropertDataSet.index,index= PropertDataSet["_id"])


def priceRecommendations(ids):
    roomsList=[]
    for i in ids:
        indexes = indice[i]
    priceSigScore = list(enumerate(priceSig[indexes]))
    priceSigScore = sorted(priceSigScore,key=lambda x: x[1],reverse=True)
    recomIndex = [i[0] for i in priceSigScore]
    roomsList.append(PropertDataSet["_id"].iloc[recomIndex])



    return roomsList

def tagsRecommendation(ids):
    roomsList = []
    for i in ids:
        indexes = indice[i]

    tagsSigScore = list(enumerate(tagsSig[indexes]))

    tagsSigScore = sorted(tagsSigScore, key=lambda x: x[1], reverse=True)

    recomIndex = [i[0] for i in tagsSigScore]

    roomsList.append(PropertDataSet["_id"].iloc[recomIndex])

    return roomsList





def bedroomRecommendation(ids):
    roomsList = []
    for i in ids:
        indexes = indice[i]

    bedSigScore = list(enumerate(bedSig[indexes]))

    bedSigScore = sorted(bedSigScore, key=lambda x: x[1], reverse=True)

    recomIndex = [i[0] for i in bedSigScore]

    roomsList.append(PropertDataSet["_id"].iloc[recomIndex])

    return roomsList


def bathRecommendation(ids):
    roomsList=[]
    for i in ids:
        indexes = indice[i]

    bathSigScore = list(enumerate(bathSig[indexes]))

    bathSigScore = sorted(bathSigScore,key=lambda x: x[1],reverse=True)

    recomIndex = [i[0] for i in bathSigScore]


    roomsList.append(PropertDataSet["_id"].iloc[recomIndex])



    return roomsList



@app.route("/GET/<id>")




def index(id):
    user = []
    for x, y in zip(UserDataSet["_id"], UserDataSet["userSaveHomes"]):
        if (x == id):
            user.append(y)


    user=user[0]
    data = [priceRecommendations(user), tagsRecommendation(user), bedroomRecommendation(user), bathRecommendation(user)]
    print(data)
    print("Recommendation based on price ::")

    print(priceRecommendations(user))

    print("Recommendation on basis of searched tags ::")
    print(tagsRecommendation(user))

    print("Recommendation on basis of number of bedrooms ::")

    print(bedroomRecommendation(user))

    print("Recommendation on basis of number of bathroom")

    print(bathRecommendation(user))

    return str(data)

#ids=index(id)


#user=index(id)



PropertDataSet =pd.DataFrame( pd.read_json("P:/Data Science/webmob/Property.json"))

UserDataSet =pd.DataFrame( pd.read_json("P:/Data Science/webmob/User.json"))


PropertDataSet = PropertDataSet.loc[:,["_id","bedroom","bathroom","tags","price"]]

PropertDataSet["tags"] = PropertDataSet["tags"].astype(str)

uid="5d8db34ae19650e22b29cc5a"



@app.route("/GET/<id>/summary")
def preference(id):
    userDetails={}
    print("---------------------------------------------------------------------------------------------------------------")
    print("-----------------------------------My Contact details----------------------------------------------------------")
    details=[]
    cnt=0
    userHomes=[]
    tags=[]
    for ids,phone,mail,fname,tag in zip(UserDataSet["_id"],UserDataSet["phone"],UserDataSet["email"],UserDataSet["name"].values,UserDataSet["userSaveHomes"]):
        if(id==ids):
            userDetails["phone"]=phone
            userDetails["mail"]=(mail)
            userDetails["name"]=(fname['first'])+" "+(fname['last'])
            userHomes.append(tag)
            print("Name ::",userDetails["name"],"\tMail id ::",userDetails["mail"])
            print("Phone number ::",userDetails["phone"])
    bedrooms=[]
    bathroom=[]
    for i in userHomes[0]:
        for ids,no,bno,tag in zip(PropertDataSet["_id"],PropertDataSet["bedroom"],PropertDataSet["bathroom"],PropertDataSet["tags"]):
            if(ids==i):
                bedrooms.append(no)
                bathroom.append(bno)
                tags.append(tag)
    print("---------------------------------------------------------------------------------------------------------------")
    print(("-----------------------------My preference :-----------------------------------------------------------------"))

    print("Bedroom range ::",min(bedrooms),"-",max(bedrooms),"\t Bathroom range ::",min(bathroom),"-",max(bathroom))
    print("---------------------------------------------------------------------------------------------------------------")
    print("--------------------------------------My Searched tags ::------------------------------------------------------")
    print(tags)
    dip="See terminal for output"
    
    return dip


if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```
